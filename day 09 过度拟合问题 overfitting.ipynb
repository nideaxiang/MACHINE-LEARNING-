# 解决过拟合

- 方法一：收集更多的数据来解决波动
- 方法二：能否使用更少的特征项
       选择最合适的功能集有时也成为功能选择
- 方法三：正则化：相比于直接将特征项抹除，采用更温和的方式：将其参数值变得很小。
正则化的作用 是它可以让你保留所有特征， 但它们只是防止特征 产生过大的影响

## 正则化的成本函数：
惩罚所有的w，在原本的成本函数后加一个正则化的项，该项具有很大的系数，所以当成本函数需要求得最小值时，需要将正则化的项的w变得很小
