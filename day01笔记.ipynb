## 2. 什么是机器学习
    来自亚瑟·塞缪尔的机器学习的定义。 她将机器学习定义 为使 计算机能够在不 经过明确编程的情况下进行学习的研究领域
    三种主要的类型：
    1.有监督学习（主要）
    2.无监督学习
    3.优化推荐系统
